```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)

#seeing what column heads are
purchase_data.head()

```

## Player Count

* Display the total number of players


```python
Total_Players = purchase_data.SN.count()
PlayerCount = pd.DataFrame([{"Total Players": Total_Players}])
PlayerCount
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Unique_Items = purchase_data["Item ID"].nunique()
Average_Price = purchase_data["Price"].mean()
Number_Purchase = purchase_data["Item ID"].count()
Total_Purchase = purchase_data["Price"].sum()

#creates dataframe
Purchase = pd.DataFrame([
    {"Unique Items": Unique_Items, "Average Price": Average_Price,
     "Number of Purchases": Number_Purchase, "Total Purchases": Total_Purchase }
])

#formats currency
Purchase["Average Price"] = Purchase["Average Price"].map('${:,.2f}'.format)
Purchase["Total Purchases"] = Purchase["Total Purchases"].map('${:,.2f}'.format)

#prints out table
Purchase
```

## Gender Demographics

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
GenderTable = purchase_data[["Gender", "SN"]]
GenderOnly = GenderTable.groupby(['Gender']).count()
#format
GenderOnly["Percentage of Players"] = (100*GenderOnly["SN"]/Total_Players).map('%{:,.2f}'.format)
GenderOnly = GenderOnly.rename(columns={'SN': 'Total Count'})
GenderOnly
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, etc. by gender


* For normalized purchasing, divide total purchase value by purchase count, by gender


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
TotalPurchase = purchase_data.groupby('Gender')

TotalCount = TotalPurchase['Item ID'].count()

TotalPrice = TotalPurchase['Price'].sum()
TotalPrice = TotalPrice.map('${:,.2f}'.format)

AveragePrice = TotalPurchase['Price'].mean()
AveragePrice = AveragePrice.rename(columns={})
AveragePrice = AveragePrice.map('${:,.2f}'.format)

NormPrice= (TotalPurchase['Price'].sum()).divide(TotalPurchase['Item ID'].count())
NormPrice = NormPrice.map('${:,.2f}'.format)

TableData= pd.concat([TotalCount, TotalPrice, AveragePrice,NormPrice], axis=1)
TableData = TableData.rename(columns={'Item ID': 'Purchase Count', 'Price':'Total Price',
                                      0:'Average Price', 1:'Normalized Price'
                                     })
TableData 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

bins = pd.cut(purchase_data['Age'],age_bins)

AgeCount = purchase_data.groupby(bins)['Age'].count()
AgeCount = AgeCount.rename(columns={})

AgePercentage = (AgeCount/Total_Players) * 100
AgePercentage = AgePercentage.map('%{:.2f}'.format)

AgeDem = pd.concat([AgeCount, AgePercentage],axis=1)
AgeDem.index = group_names
AgeDem = AgeDem.rename(columns={0: 'Total', 1: 'Percentage'})
AgeDem
```
